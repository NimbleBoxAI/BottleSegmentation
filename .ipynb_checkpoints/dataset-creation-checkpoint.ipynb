```python
from tqdm import tqdm
```

```python
ids = []
labels = []
for i in tqdm(range(len(df_g))):
    x = df_g['image_id'].iloc[i]
    path = f'../archive/celebA/archive/img_align_celeba/img_align_celeba/{x}'
    #path = f'../../test/1/face-{i+1}.png'
    image = cv2.imread(path)[..., ::-1].astype(np.float32)
    image = cv2.resize(image, (384, 384))
    image = image / 255
    image = normalize_fn(image).float()[None].cuda()
        # run head for glasses feature
    output = model(image, 'glasses')
    ids.append(x)
    labels.append(torch.argmax(output).detach().cpu().numpy())
```

```python
celeba = pd.DataFrame({'id':ids, 'glasses':labels})
```
